```python
base=[]
for day in list(set(red["dia"])):
    peso_par=red[red["dia"]==day].groupby(['sourcecol', 'targetcol']).count()["daycol"].to_dict()

    base.append([(k[0], k[1], v, day) for k,v in peso_par.items()])
            
```

```python
base=[x for i in base for x in i]
```

```python
red2=pd.DataFrame(base).rename(columns={0:"sourcecol",1: "targetcol",2: "weight",3: "day"})
```

```python
red2["day"]=pd.to_datetime(red2["day"], dayfirst=True)
```

```python
red2=red2.sort_values("day")
```
